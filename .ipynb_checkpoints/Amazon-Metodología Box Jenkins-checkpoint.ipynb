# Estimación DF con constante y tendencia

```python
test_DF = sm.tsa.stattools.adfuller(amzn.Close.dropna(), regression='ct')
print('ADF Statistic: %f' % test_DF[0])
print('p-value: %f' % test_DF[0])
print('Critical Values:')
for key, value in test_DF[4].items():
    print('\t%s: %.3f' % (key,value))
    
if test_DF[0] < test_DF[4]['5%']:
    print('Reject Ho - Time series is Stationary')
else:
    print('Failed to Reject Ho - Times series Non-Stationary')
```

# Estimación DF sin tendencia

```python
test_DF = sm.tsa.stattools.adfuller(amzn.Close.dropna(), regression='c')
print('ADF Statistic: %f' % test_DF[0])
print('p-value: %f' % test_DF[0])
print('Critical Values:')
for key, value in test_DF[4].items():
    print('\t%s: %.3f' % (key,value))
    
if test_DF[0] < test_DF[4]['5%']:
    print('Reject Ho - Time series is Stationary')
else:
    print('Failed to Reject Ho - Times series Non-Stationary')
```

# Se concluye desde un inicio que es estacionario, por lo tanto se procede a evaluar el modelo la FAS y FAP

```python
fig,(ax1,ax2) = plt.subplots(1,2,figsize=(12,4))
ax1.set_xlim([0,15])
ax2.set_xlim([0,15])
ax1.set_ylim([-.1,.1])
ax2.set_ylim([-.1,.1])
# ACF chart
fig = sm.graphics.tsa.plot_acf(amzn.pct_ch.dropna(), ax=ax1)
# PACF chart
fig = sm.graphics.tsa.plot_pacf(amzn.pct_ch.dropna(), ax=ax2)
# ignorar el primero
```

```python
# Vemos presencia de un AR(2) puro
# en python la librería statsmodels solo permite 
# modelar el número de rezagos de manera continua
```

```python
mod = ARMA(amzn.pct_ch.dropna(), order=(2,0))
result = mod.fit()
print(result.summary())
```

```python
predictions = pd.Series(result.fittedvalues, copy=True)
# convert to cumulative sum
predictions_sum = predictions.cumsum()
values_plus = amzn.Close.add(predictions_sum)
plt.figure(figsize=(12,10))
plt.plot(amzn.Close)
plt.plot(values_plus,color='red')
```

```python
result.plot_predict(5000,5140)
```

```python
plt.subplots(1,1,figsize=(12,4))
plt.plot(amzn.Close[5000:5131])
plt.subplots(1,1,figsize=(12,4))
plt.plot(amzn.pct_ch[5000:5131])
```

# Es un AR(2) puro

```python
fig,(ax1,ax2) = plt.subplots(1,2,figsize=(12,4))
ax1.set_xlim([0,15])
ax2.set_xlim([0,15])
# ACF chart
fig = sm.graphics.tsa.plot_acf(result.resid.dropna(), ax=ax1)
# PACF chart
fig = sm.graphics.tsa.plot_pacf(result.resid.dropna(), ax=ax2)
# Vemos que los residuos tienen un comportamiento normal
```

```python
plt.plot(result.resid)
# Al analizar la serie de residuos generados por el modelo, podemos ver que el gráfico
# nos suguere que la serie sigue un proceso de 
# heterocedasticidad condicional autorregresivo
```

# Es claro que precenta un problema de heterocedasticidad condicionada

```python
# entonces para confirmar vermos el comportamiento de los residuos al cuadrado
plt.plot(np.square(result.resid));
```

```python
# Confirmamos con los gráficos de ACF y PACF

fig,(ax1,ax2) = plt.subplots(1,2,figsize=(12,4))
ax1.set_xlim([0,15])
ax2.set_xlim([0,15])
ax1.set_ylim([-.1,.2])
ax2.set_ylim([-.1,.2])
# ACF chart
fig = sm.graphics.tsa.plot_acf(np.square(result.resid.dropna()), ax=ax1)
# PACF chart
fig = sm.graphics.tsa.plot_pacf(np.square(result.resid.dropna()), ax=ax2)
# No solo se observa un posible modelo ARCH, si no que además, el número de rezagos
# a incluir sería bastante elevado.
# procederemos a evaluar un GARCH
```
